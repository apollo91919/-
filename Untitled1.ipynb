```python
import numpy as np
import matplotlib.pyplot as plt

# 自訂輸入條件
L = float(input("請輸入棒材長度 L: "))
T = float(input("請輸入模擬總時間 T: "))
h = float(input("請輸入空間步長 h: "))
dt = float(input("請輸入時間步長 dt: "))

N = int(L / h)  # 空間步數
M = int(T / dt)  # 時間步數
alpha = 0.01  # 熱擴散係數
A = alpha * dt / (h**2)

# 初始化溫度場
U = np.zeros((N+1, M+1))

# 初始條件
for i in range(N+1):
    U[i, 0] = 4 * i * h * (L - i * h)

# 邊界條件
U[0, :] = 0
U[N, :] = 0

# 數值求解
for k in range(M):
    for i in range(1, N):
        U[i, k+1] = A * U[i+1, k] + (1 - 2*A) * U[i, k] + A * U[i-1, k]

# 繪製溫度分佈圖
x = np.linspace(0, L, N+1)
for t in range(0, M+1, M//5):
    plt.plot(x, U[:, t], label=f't={t*dt:.2f}')

plt.xlabel('x')
plt.ylabel('u(x,t)')
plt.title('溫度分佈')
plt.legend()
plt.show()

# 溫度等高線圖
extent = [0, T, 0, L]
levels = np.arange(0, np.max(U), 0.1)
plt.contourf(U, levels, origin='lower', extent=extent, cmap=plt.cm.jet)
plt.colorbar()
plt.xlabel('時間 (t)')
plt.ylabel('位置 (x)')
plt.title('溫度等高線圖')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 自定义输入条件
L = float(input("请输入棒材长度 L (建议范围 1.0 到 10.0): "))
T = float(input("请输入模拟总时间 T (建议范围 0.1 到 10.0): "))
h = float(input("请输入空间步长 h (建议范围 0.1 到 1.0): "))
dt = float(input("请输入时间步长 dt (建议范围 0.0001 到 0.01): "))

N = int(L / h)  # 空间步数
M = int(T / dt)  # 时间步数
alpha = 0.01  # 热扩散系数
A = alpha * dt / (h**2)

# 检查稳定性条件
if dt > h**2 / (2 * alpha):
    raise ValueError("时间步长 dt 不满足数值稳定性条件，请减小 dt 或增大 h。")

# 初始化温度场
U = np.zeros((N+1, M+1))

# 初始条件
for i in range(N+1):
    U[i, 0] = 4 * i * h * (L - i * h)

# 边界条件
U[0, :] = 0
U[N, :] = 0

# 数值求解
for k in range(M):
    for i in range(1, N):
        U[i, k+1] = A * U[i+1, k] + (1 - 2*A) * U[i, k] + A * U[i-1, k]

# 绘制不同时间的温度随空间坐标的变化
x = np.linspace(0, L, N+1)
time_steps = [0, int(M*0.3), int(M*0.6), int(M*0.9), M]

for t in time_steps:
    plt.plot(x, U[:, t], label=f't={t*dt:.2f}', linewidth=1.0)

plt.xlabel('x', fontsize=20)
plt.ylabel('u(x,t)', fontsize=20)
plt.xlim(0, L)
plt.ylim(-2, np.max(U) + 2)
plt.legend(loc='upper right')
plt.title('不同时间的温度分布')
plt.show()

# 温度等高线图
extent = [0, T, 0, L]
levels = np.arange(0, np.max(U), 0.1)
plt.contourf(U, levels, origin='lower', extent=extent, cmap=plt.cm.jet)
plt.colorbar()
plt.xlabel('时间 (t)', fontsize=20)
plt.ylabel('位置 (x)', fontsize=20)
plt.title('温度等高线图')
plt.show()
```
